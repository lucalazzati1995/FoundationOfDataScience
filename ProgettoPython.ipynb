Convert the app sizes to a number

```python
# 1.Converti le dimensioni delle app in un numero
df['Size'].str.extractall(r'([a-z])', flags = re.IGNORECASE).count() #Conto i valori che contengono lettere.
```

```python
df['Size'].str.extractall(r'([kM]$)').count() 
#Le due righe combaciano perfettamente quindi ho verificato che tutti i valori finiscono per M o per k
```

```python
Kilo= df['Size'].str.extractall(r'(k$)').index.get_level_values(0) #metto a 0 perchè è un MultiIndex.
# è un oggetto Index contenente tutti gli Indici
```

```python
df['Size'].replace(regex=True, inplace=True, to_replace=r'\D$', value=r'') #elimino le righe con i valori M e k
```

```python
df['Size'].str.extractall(r'([a-z])', flags = re.IGNORECASE) #verifico che in tutte le righe non vi siano più lettere
```

```python
df['Size']=pd.to_numeric(df['Size']) # convertire tutti i valori in float per avere la stessa scala rispetto a MB
```

```python
df["Size"].loc[Kilo] = (df["Size"].loc[Kilo]/1000) # conversione da kb a mb e divisione per mille
df.rename(columns={'Size': 'Size(MB)'}, inplace=True) #ora che i valori sono in megabyte, rinomino la colonna
df.head() # Visualizzazione testa
```

Create two dataframes: one for the genres and one bridging apps and genres. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
# 8.Crea due dataframe: uno per i generi e un app e generi per il bridging. 
#Così, ad esempio, l'app Pixel Draw - Number Art Coloring Book appare due volte nella tabella di bridging, 
#una volta per Art & Design, una volta per Creativity 

df[df['App']=='Pixel Draw - Number Art Coloring Book']
```

```python
app_gene = df[['App', 'Genres']] #Creo un dataframe con solo le due colonne che mi interessano

app_gene.head(20) #mostra il nuovo dataframe
```

```python
genres =app_gene["Genres"].str.split(";", expand = True) #creo un dataframe con due Colonne. Una per ogni genere

genres
```

```python
app_gene['Genre'] = genres[0] #sostituisco la colonna 'Genre' (che aveva le coppie di generi) con la colonna del primo genere per ogni app

app_gene['Genre']
```

```python
secondgenre = pd.DataFrame({'App': app_gene['App'],'Genre':genres[1]}) #creo un dataframe che ha le app in corrispondenza del loro secondo genere. Le app che non hanno il secondo genere hanno un missing value

secondgenre
```

```python
app_gene.drop(columns = ["Genres"], inplace = True) #elimino la colonna genres di partenza dal app_gene
```

```python
secondgenre.dropna(subset = ['Genre'], inplace=True) #elimino i missing values per le app che non hanno il secondo genere 
```

```python
dfconcat=pd.concat([app_gene, secondgenre], ignore_index=True) #creo il dataframe finale che ha tutte le app in corrispondenza di un genere. Le app che hanno due generi compaiono due volte. Aggiungo ignore_index=True, sennò in corrispondenza dell'istanza 4 ci sarebbero state due Pixel Draw, una per ogni genere. Invece ora la prima è sulla riga 4, la seconda è sulla riga 10839
```

For each genre, create a new column of the original dataframe. The new columns must have boolean values (True if the app has a given genre)

```python
# 9.Per ogni genere, crea una nuova colonna del dataframe originale. 
#Le nuove colonne devono avere valori booleani (Vero se l'app ha un determinato genere) 

df=df.join(genres) #aggiungo al dataframe le due colonne genere1 e genere 2 che avevo creato.
```

```python
df['Price'].value_counts()
```

```python
df['Price']=df['Price'].replace(r"\$", value = "", regex = True) #faccio questa operazione per togliere il $
df["Price"] = pd.to_numeric(df["Price"]) # da stringa a numero
```
