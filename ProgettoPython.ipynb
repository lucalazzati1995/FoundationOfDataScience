## Convert the app sizes to a number

```python
# 1.Converti le dimensioni delle app in un numero
df['Size'].str.extractall(r'([a-z])', flags = re.IGNORECASE).count() #Conto i valori che contengono lettere.
```

```python
df['Size'].str.extractall(r'([kM]$)').count() 
#Le due righe combaciano perfettamente e ho verificato che tutti i valori finiscono per M o per k
```

```python
Kilo= df['Size'].str.extractall(r'(k$)').index.get_level_values(0) #metto a 0 perchè è un MultiIndex.
Kilo # è un oggetto Index contenente tutti gli Indici
```

```python
df['Size'].replace(regex=True, inplace=True, to_replace=r'\D$', value=r'') #elimino le righe con i valori M e k
df['Size']
```

```python
df['Size'].str.extractall(r'([a-z])', flags = re.IGNORECASE) #verifico che in tutte le righe non vi siano più lettere
```

```python
df['Size']=pd.to_numeric(df['Size']) # convertire tutti i valori in float per avere la stessa scala rispetto a MB
df['Size']
```

```python
df["Size"].loc[Kilo] = (df["Size"].loc[Kilo]/1000) # conversione da kb a mb e divisione per mille
df["Size"]
```

```python
df.rename(columns={'Size': 'Size(MB)'}, inplace=True) #ora che i valori sono in megabyte, rinomino la colonna
df.head() # Visualizzazione testa
```

## Create two dataframes: one for the genres and one bridging apps and genres. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
# 8.Crea due dataframe: uno per i generi e un app e generi per il bridging. 
#Così, ad esempio, l'app Pixel Draw - Number Art Coloring Book appare due volte nella tabella di bridging, 
#una volta per Art & Design, una volta per Creativity 

df[df['App']=='Pixel Draw - Number Art Coloring Book']
```

```python
df1 = df[['Genres','App']] #Creo un dataframe con solo le due colonne che mi interessano

df1.head(20) #mostra il nuovo dataframe
```

```python
genres =df1["Genres"].str.split(";", expand = True) #creo un dataframe con due Colonne. Una per ogni genere

genres
```

```python
df1['Genre'] = genres[0] #sostituisco la colonna 'Genres' (che aveva le coppie di generi) con la colonna del primo genere per ogni app

df1['Genre']
```

```python
secondgenre = pd.DataFrame({'App': df1['App'],'Genre':genres[1]}) #creo un dataframe che ha le app in corrispondenza del loro secondo genere.
#Le app che non hanno il secondo genere hanno un missing value

secondgenre
```

```python
df1.drop(columns = ['Genres'], inplace = True) #elimino la colonna genres di partenza dal app_gene
```

```python
secondgenre.dropna(subset = ['Genre'], inplace=True) #elimino i missing values per le app che non hanno il secondo genere 
```

```python
dfconcat=pd.concat([df1, secondgenre], ignore_index=True) 

dfconcat
#creo il dataframe finale che ha tutte le app in corrispondenza di un genere.
#Le app che hanno due generi compaiono due volte. 
#Aggiungo ignore_index=True, sennò in corrispondenza dell'istanza 4 ci sarebbero state due Pixel Draw, una per ogni genere.
```

## For each genre, create a new column of the original dataframe. The new columns must have boolean values (True if the app has a given genre)

```python
# 9.Per ogni genere, crea una nuova colonna del dataframe originale. 
#Le nuove colonne devono avere valori booleani (Vero se l'app ha un determinato genere) 

df=df.join(genres) #aggiungo al dataframe le due colonne genere1 e genere 2 che avevo creato.
```

```python
df.rename(columns={0: 'Genre1', 1: 'Genre2'}, inplace=True)
```

```python
df = pd.concat([df, pd.get_dummies(df['Genre1']).astype(bool), pd.get_dummies(df['Genre2']).astype(bool)], axis=1) #creo le colonne binarie per il primo genere e per il secondo genere.
```

## For each genre, compute the average rating. What is the genre with highest average?

```python
# 10.Per ogni genere, calcola la valutazione media. Qual è il genere con la media più alta? 
dfgenere = df.merge(dfconcat, on='App') #creo un nuovo datframe che ha che ha la stessa app che compare due volte (solo per le app con due generi), una volta per genere.
dfgenere.head()
```

```python
dfgenere.groupby('Genre')['Rating'].mean()
```

```python
print("The average of the highest rating corresponds to the genres is " +dfgenere.groupby('Genre')['Rating'].mean().argmax() + " ,with a value of " + str(dfgenere.groupby('Genre')['Rating'].mean().max()))
```
